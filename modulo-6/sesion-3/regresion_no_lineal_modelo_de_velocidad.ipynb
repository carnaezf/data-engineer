# Visualización de un modelo de regresión no lineal
##  Interpretación del modelo cuadrático de velocidad

###  Modelo propuesto

**Basándonos en observaciones reales, podemos modelar esta relación así:**

$$
v = -0.5 \cdot t^2 + 6 \cdot t
$$

---

###  Interpretación de la fórmula en lenguaje natural

**¿Qué nos está diciendo esta ecuación?**

- **\( v \)**: velocidad del auto en kilómetros por hora (km/h) → **es lo que queremos predecir**.
- **\( t \)**: tiempo transcurrido en horas (h) → **es lo que conocemos**.

- El coeficiente **\(-0.5\)** delante de \( t^2 \) indica que, a medida que pasa el tiempo, la **velocidad deja de aumentar y comienza a disminuir**.  
  Esto es típico en situaciones reales donde existen **resistencia, pendiente o fatiga del motor**.

- El coeficiente **\(+6\)** delante de \( t \) nos dice que al inicio hay una **aceleración fuerte**: por cada hora, el auto gana velocidad rápidamente.

---

###  En resumen

> **El auto acelera al principio, alcanza una velocidad máxima, y luego desacelera.**

---

###  Cálculo concreto con valores reales

**• Si han pasado 1 hora:**

$$
v = -0.5(1)^2 + 6(1) = -0.5 + 6 = \boxed{5.5\text{ km/h}}
$$

---

**• Si han pasado 5 horas:**

$$
v = -0.5(5)^2 + 6(5) = -12.5 + 30 = \boxed{17.5\text{ km/h}}
$$

---

**• Si han pasado 10 horas:**

$$
v = -0.5(10)^2 + 6(10) = -50 + 60 = \boxed{10\text{ km/h}}
$$

---

###  ¿Qué nos dicen estos resultados?

 **El modelo nos indica que:**

-  A la **hora 1**, el auto aún está **acelerando**.
-  A la **hora 5**, alcanza su **velocidad máxima**.
-  A la **hora 10**, ya está en **fase de desaceleración**.


```python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt

# Definir el dominio (tiempo en horas)
t = np.linspace(0, 12, 100)
# Definir la función de velocidad
v = -0.5 * t**2 + 6 * t

# Puntos destacados para análisis
tiempos_especiales = [1, 5, 10]
velocidades_especiales = [-0.5 * ti**2 + 6 * ti for ti in tiempos_especiales]

# Crear gráfico
plt.figure(figsize=(10, 6))
plt.plot(t, v, label='v = -0.5t² + 6t', color='blue')
plt.scatter(tiempos_especiales, velocidades_especiales, color='red', zorder=5)

# Anotar los puntos
for t_i, v_i in zip(tiempos_especiales, velocidades_especiales):
    plt.annotate(f"t = {t_i}h\nv = {v_i:.1f} km/h",
                 xy=(t_i, v_i), xytext=(t_i+0.3, v_i+2),
                 arrowprops=dict(arrowstyle="->", color='black'))

# Configurar gráfico
plt.title('Modelo de velocidad en función del tiempo (Regresión No Lineal)')
plt.xlabel('Tiempo (horas)')
plt.ylabel('Velocidad (km/h)')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 🧪 CÁLCULO MANUAL: Prueba tus propios valores

# Ingresar valores de tiempo
t1 = 1
t2 = 5
t3 = 10

# Calcular velocidad manualmente
v1 = -0.5 * t1**2 + 6 * t1
v2 = -0.5 * t2**2 + 6 * t2
v3 = -0.5 * t3**2 + 6 * t3

# Mostrar resultados
print("Cálculo manual de la velocidad usando el modelo v = -0.5t² + 6t\n")
print(f"🔴 Si han pasado {t1} hora(s): v = {v1} km/h")
print(f"🔴 Si han pasado {t2} hora(s): v = {v2} km/h")
print(f"🔴 Si han pasado {t3} hora(s): v = {v3} km/h")

# Mensaje de cierre
print("\n Compara estos resultados con los puntos rojos del gráfico anterior.\n")
print("¿Coinciden? ¡Excelente! Así se ve en acción un modelo de regresión no lineal.")
```
